```python
petitioner_dict = {}
for line in petitioner_map.split('\n'):
    key, value = line.split('\t')
    petitioner_dict[int(key)] = value[:50]
```

```python
respondent_dict = {}
for line in respondent_map.split('\n'):
    key, value = line.split('\t')
    respondent_dict[int(key)] = value[:50]
```

```python
winner_dict = {0: 'respondent',
               1: 'petitioner'}
```

```python
df = df.dropna(subset=['petitioner'], axis=0)
df['petitioner'] = df['petitioner'].astype(int)
df['petitioner'] = df['petitioner'].map(petitioner_dict)
```

```python
df = df.dropna(subset=['respondent'], axis=0)
df['respondent'] = df['respondent'].astype(int)
df['respondent'] = df['respondent'].map(respondent_dict)
```

```python
df = df.dropna(subset=['partyWinning'], axis=0)
df['partyWinning'] = df['partyWinning'].astype(int)
df = df[df['partyWinning'] < 2]
df['partyWinning'] = df['partyWinning'].map(winner_dict)
```

```python
df = df[['dateDecision', 'justiceName', 'caseName', 'petitioner', 'respondent', 'partyWinning']]
df['winner'] = pd.Series()
df['loser'] = pd.Series()
```

```python
df['winner'] = df.apply(lambda row: row['petitioner']
                                    if row['partyWinning'] == 'petitioner'
                                    else row['respondent'],
                        axis=1)
```

```python
df['loser'] = df.apply(lambda row: row['petitioner']
                                   if row['partyWinning'] == 'respondent'
                                   else row['respondent'],
                        axis=1)
```

```python
# Two parties from same industry provides no preference info. Cut.
df = df[df['petitioner'] != df['respondent']]
```

```python
justices = df['justiceName'].unique()
dataframes = [df.groupby('justiceName').get_group(justice)
              for justice
              in justices]

def get_affinity(df):
    # Each petitioner or respondent appearance is an opportunity
    # to find for that industry.
    res_opp = df['petitioner'].value_counts()
    pet_opp = df['respondent'].value_counts()
    opportunities =  res_opp.add(pet_opp, fill_value=0)
    # Only if n >= 10
    opportunities = opportunities[opportunities.values >= 10]
    percentages = df['winner'].value_counts() / opportunities
    return percentages
    
    
affinity = {j: get_affinity(df)
            for j, df
            in zip(justices, dataframes)}

results_df = pd.DataFrame(columns=['justice', 'rank', 'party', 'percentage'])

for justice, series in affinity.items():
    ts = series.sort_values(ascending=False)[0:50]
    tdf = pd.DataFrame(columns=['justice', 'rank', 'party', 'percentage'])
    tdf['rank'] = pd.Series([x for x in range(1, 51)])
    tdf['justice'] = justice
    tdf['party'] = ts.index.values
    tdf['percentage'] = ts.values
    results_df = results_df.append(tdf)
```
